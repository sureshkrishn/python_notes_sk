#### Relational Operators


```python
amy_age = 25
john_age = 30

## Equal to operator
print(amy_age == john_age)

if (amy_age == john_age) :
    print(" Both Amy and John have same age")
else :
    print(" Amy and John have difference in their age")

## Greater Than Operator
print(john_age > amy_age)

## Greater than or equal to
print(john_age >= amy_age)

## less Than Operator
print(john_age < amy_age)

## less than or equal to
print(john_age <= amy_age)

## Not equal to
print(john_age != amy_age)
```

* False or True = True
* False or False = False

```python
## Displaying bus fare based on age group
age = int(input("Enter the age :"))

if (0 <= age < 3) :
    print("There is no bus fare for age group less than 3")
elif (3 <= age < 10) :
    print("The bus fare for age between 3 and 9 is 20")
elif (10 <= age < 20) :
    print("The bus fare for age between 10 and 19 is 25")
else :
    print("The bus fare for age greater than 19 is 30")
```

```python
## Finding the largest number among the three numbers using nested if else
num1, num2, num3 = 96, 56, 48

print("The given numbers are : ", num1, num2, num3)
if (num1 > num2) :
    if (num1 > num3) :
        print("The largest number is ", num1)
    else :
        print("The largest number is ", num3)

else :
    if (num2 > num3) :
        print("The largest number is ", num2)
    else :
        print("The largest number is ", num3)
            
```

```python
## While Loop
num = 1000
i = 1
total = 0

while (i <= num) :
    print('I = ', i)
    total += i
    print('Total = ', total)
    i += 1
    

## Printing the total value
print("The sum of number between 1 and 10 is ",total)
```
